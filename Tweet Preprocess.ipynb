```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv(" #RIPKobe4.4.csv")
data.head()
```

```python
text = data["text"]
```

```python
import preprocessor as p
```

```python
def preprocess_tweet(row):
    text = row['text']
    text = p.clean(text)
    return text
```

```python
data['text'] = data.apply(preprocess_tweet, axis=1)
```

```python
data.head()
```

```python
from gensim.parsing.preprocessing import remove_stopwords
```

```python
def stopword_removal(row):
    text = row['text']
    text = remove_stopwords(text)
    return text
```

```python
data['text'] = data.apply(stopword_removal, axis=1)
```
