```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("KobeDataClean.csv")

data
```

```python
text = data['text']
```

```python
text
```

```python
test = text[2]

# Create and generate a word cloud image:
wordcloud = WordCloud().generate(test)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
words = " ".join(review for review in data.text)
print ("There are {} words in the combination of all review.".format(len(words)))
```

```python
# Create stopword list:

stopwords = set(STOPWORDS)
stopwords.update(["I", "I'm", "you", "the", "and", "try", "th", "coming", "know", "KG", "Duncan", "st", "day", "come", "officially", "AD", "y'all", "night", "videos", "follow", "watch", "bean", "amp", "AR", "selling", "FF", "pair", "eBay", "Bust", "fuck", "fucking", "shit"])
```

```python
Kobe_wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(words)

# Display the generated image:
# the matplotlib way:

plt.imshow(Kobe_wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
Kobe_wordcloud.to_file("RIPKobe_wordcloud.png")
```

```python
kobe_mask = np.array(Image.open("kobe_mask.jpg"))
kobe_mask
```

```python
# Create a word cloud image
wc = WordCloud(background_color="white", mask=kobe_mask,
               stopwords=stopwords)

# Generate a wordcloud
wc.generate(words)

# show
plt.figure(figsize=[50,25])
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wc.to_file("RIPKobe_wc.png")
```

```python
bryant_mask = np.array(Image.open("kobe_iconic.jpg"))
bryant_mask
```

```python
# Create a word cloud image
bryant = WordCloud(background_color="white", mask=bryant_mask,
               stopwords=stopwords)

# Generate a wordcloud
bryant.generate(words)

# show
plt.figure(figsize=[50,25])
plt.imshow(bryant, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
bryant.to_file("RIPKobe_wc_v1.png")
```

```python
# Generate a word cloud image
color_mask = np.array(Image.open("kobe_color.jpg"))
kobe_color = WordCloud(stopwords=stopwords, background_color="white", mode="RGBA", mask=color_mask).generate(words)

# create coloring from image
image_colors = ImageColorGenerator(color_mask)
plt.figure(figsize=[25,25])
plt.imshow(kobe_color.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")

plt.show()
```

```python
kobe_color.to_file("RIPKobe_color_v1.png")
```
